```python
from pickle import TRUE
one_hot_encoded_data = pd.get_dummies(df, columns = ['Item_Fat_Content', 'Item_Type', 'Outlet_Location_Type', 'Outlet_Type'], drop_first=TRUE)
```

```python
one_hot_encoded_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Outlet_Size'] = le.fit_transform(df['Outlet_Size'])
```

```python
one_hot_encoded_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
one_hot_encoded_data['Item_Weight'] = sc.fit_transform(one_hot_encoded_data[['Item_Weight']])
one_hot_encoded_data['Item_Visibility'] = sc.fit_transform(one_hot_encoded_data[['Item_Visibility']])
one_hot_encoded_data['Item_MRP'] = sc.fit_transform(one_hot_encoded_data[['Item_MRP']])
```

```python
one_hot_encoded_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
x = one_hot_encoded_data.drop(['Item_Outlet_Sales'], axis=1)
y = one_hot_encoded_data['Item_Outlet_Sales']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
rf = RandomForestRegressor(random_state=42)
```

```python
lr.fit(x_train, y_train)
rf.fit(x_train, y_train)
```

```python
y_pred_lr = lr.predict(x_test)
y_pred_rf = rf.predict(x_test)
```

```python
def evaluate_model(y_true, y_pred):
  mae = mean_absolute_error(y_true, y_pred)
  mse = mean_absolute_error(y_true, y_pred)
  rmse =  np.sqrt(mse)
  r2 = r2_score(y_test, y_pred)
  return mae,mse, rmse, r2

# Linear regression evaluation
mae_lr, mse_lr, rmse_lr,r2_lr = evaluate_model(y_test, y_pred_lr)
print("Linear Regression:")
print(f"MAE: {mae_lr:.2f}, MSE: {mse_lr:.2f}, RMSE: {rmse_lr:.2f}, R2 Score: {r2_lr:.2f}")

# Random Forest Evaluation
mae_rf, mse_rf, rmse_rf, r2_rf = evaluate_model(y_test, y_pred_rf)
print("\nRandom Forest:")
print(f"MAE: {mae_rf:.2f}, MSE: {mse_rf:.2f}, RMSE: {rmse_rf:.2f}, R2 Score: {r2_rf:2f}")

```
